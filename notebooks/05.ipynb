```python
df = pd.read_csv('../data/day_05_test.txt',
                 names=["x1", 'y1', 'x2', 'y2'])
```

```python
df.head(10)
```

```python
# Remove diagnaol lines
df = df.query('x1 == x2 or y1 == y2')
df.head(10)
```

```python
def get_range(a, b):
    if a < b:
        return  [a for a in range(a, b+1)] 
    elif a > b:
        return [x for x in range(b, a+1)]
    else:
        return [a]
```

```python
df.loc[:, 'x_points'] = df.apply(lambda row: get_range(row.x1, row.x2), axis=1)
df.loc[:, 'y_points'] = df.apply(lambda row: get_range(row.y1, row.y2), axis=1)
```

```python
df.head()
```

```python
df.loc[:, 'points'] = df.apply(lambda row: list(product(row.x_points, row.y_points)), axis=1)
```

```python
df.head()
```

```python
# Unpack the list of tuples to create a new column

def to_1D(series):
 return pd.Series([x for _list in series for x in _list])

to_1D(df.loc[:, 'points'])[:10]
```

```python
# Create a frequency table of the points
output = to_1D(df.loc[:, 'points']).value_counts().rename_axis('unique_values').reset_index(name='counts')
output.head(10)
```

```python
# Find the number of intersections greater than 1
output[output.counts > 1].shape[0]
```

## Part 2

```python
df = pd.read_csv('../data/day_05_test.txt',
                 names=["x1", 'y1', 'x2', 'y2'])

df.head(10)
```

```python
# # Remove diagnaol lines
# df = df.query('x1 == x2 or y1 == y2')
# df.head(10)
```

```python
def get_range(x, y):
    if x < y:
        return  [x for x in range(x, y+1)] 
    elif x > y:
        return [x for x in range(y, x+1)]
    else:
        return [x]
    
df.loc[:, 'x_points'] = df.apply(lambda row: get_range(row.x1, row.x2), axis=1)
df.loc[:, 'y_points'] = df.apply(lambda row: get_range(row.y1, row.y2), axis=1)

# Get all points along the lines
df.loc[:, 'points'] = df.apply(lambda row: list(product(row.x_points, row.y_points)), axis=1)

df.head(10)
```

```python
def to_1D(series):
 return pd.Series([x for _list in series for x in _list])


# Create a frequency table of the points
output = to_1D(df.loc[:, 'points']).value_counts().rename_axis('unique_values').reset_index(name='counts')

# Find the number of intersections greater than 1
output[output.counts > 1].shape[0]
```
